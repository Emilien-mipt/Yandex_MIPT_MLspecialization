# Экспоненциальное распределение
## Гистограмма выборки и теоретическая плотность распределения

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm
%matplotlib inline
```

```python
exp_rv = expon()
sample = exp_rv.rvs(1000)
```

```python
x = np.linspace(0,10,1000)
pdf = expon.pdf(x)
```

```python
plt.hist(sample, bins = 20, density=True, label='sample')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
```

## Оценка распределения выборочного среднего

### n = 5

```python
n = 5
sample = np.array([])

i = 1
while i <= 1000:
    sample_i = exp_rv.rvs(n)
    mean_i = sum(sample_i) / n
    sample = np.append(sample,mean_i)
    i+=1
    
plt.hist(sample, bins = 20, density=True, label='mean')

x = np.linspace(0,10,1000)
normal = norm(exp_rv.mean(),exp_rv.std())
pdf = normal.pdf(x)
plt.plot(x, pdf, label='normal pdf', alpha=0.5)
```
